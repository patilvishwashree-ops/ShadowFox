  # LIBRARY 1: Matplotlib

ðŸ”¹ What is Matplotlib?

          Matplotlib is a basic Python visualization library used to create different types of graphs.

ðŸ”¹ Why Matplotlib exists

          Python works with numbers, not visuals
          Matplotlib converts numerical data into graphical representations

ðŸ”¹Where to use Matplotlib

Use Matplotlib when:

           You want full control over the appearance of graphs
           You are performing basic data analysis
           You need simple and static visualizations

Matplotlib is commonly used in:

           Data analysis projects
           Academic research
           Reports and presentations

```python
import matplotlib.pyplot as plt           # plt is a commonly used short name for Matplotlib functions.
import seaborn as sns
import pandas as pd
import numpy as np
```

## Types of Graphs in Matplotlib
Matplotlib is a low-level Python visualization library used to create static plots with high customization.


# 1. Line Plot

What it shows:

          Displays changes or trends over time.

Where to use:

          Sales over time
          Temperature variation
          Stock price trends

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line Plot using Matplotlib")
plt.show()
```

# 2. Bar Chart

What it shows:

          Compares values across different categories.

Where to use:

            Product sales
            Student mark
            Population comparison

```python
categories = ['A', 'B', 'C', 'D']
values = [15, 25, 10, 30]

plt.bar(categories, values)
plt.title("Bar Chart using Matplotlib")
plt.show()
```

# 3. Scatter Plot

What it shows:

                Shows the relationship between two numerical variables.

Where to use:

               Height vs weight
               Study hours vs marks

```python
x = [1, 2, 3, 4]
y = [10, 15, 20, 25]

plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.show()
```

# 4. Histogram

What it shows:

               Displays the distribution of data values.

Where to use:

               Age distribution
               Exam score distribution

```python
data = [10, 20, 20, 30, 40, 40, 40, 50]

plt.hist(data, bins=5)
plt.title("Histogram Example")
plt.show()
```

# 5. Pie Chart

What it shows:

                       Represents data as proportions or percentages.

Where to use:

                      Market share
                      Budget allocation

```python
sizes = [40, 35, 25]
labels = ['A', 'B', 'C']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
```

# Library 2: Seaborn
ðŸ”¹ What is Seaborn?

        Seaborn is a Python data visualization library built on top of Matplotlib.
        Think of it like this:
        Matplotlib = foundation
        Seaborn = better design + less code



ðŸ”¹Why Seaborn exists

 Matplotlib is powerful but:

       Requires more code
       Graphs look basic by default

Seaborn:

        Reduces code
        Improves visual appearance
        Works very well with Pandas DataFrames

ðŸ”¹Where to use Seaborn

Use Seaborn when:

       You are doing Exploratory Data Analysis (EDA)
       You want good-looking plots quickly
       You are working with dataframes
       You want to understand patterns and relationships

Seaborn is commonly used in:

        Data analysis projects
        Machine learning exploration
        Reports and dashboards

# Types of Graphs in Seaborn
Seaborn automatically creates clean and attractive graphs.

# 1. Line Plot

What it shows:

       Displays trends over time.

Where to use:

     Monthly sales
     Temperature changes

```python
x = [1, 2, 3, 4]
y = [5, 10, 15, 20]

sns.lineplot(x=x, y=y)
plt.title("Seaborn Line Plot")
plt.show()

```

# 2. Bar Plot

What it shows:

     Compares values across categories.

Where to use:

      Product comparison
      Category-wise counts

```python
categories = ['A', 'B', 'C']
values = [10, 20, 15]

sns.barplot(x=categories, y=values)
plt.title("Seaborn Bar Plot")
plt.show()
```

# 3. Scatter Plot

What it shows:

         Shows relationship between two numerical variables.

Where to use:

      Height vs weight
      Study hours vs marks

```python
sns.scatterplot(x=x, y=y)
plt.title("Seaborn Scatter Plot")
plt.show()
```

# 4. Histogram

What it shows:

       Displays the distribution of data.

Special feature:

        Seaborn can add a smooth curve called KDE.

```python
data = np.random.randn(1000)

sns.histplot(data, bins=20)
plt.title("Histogram using Seaborn")
plt.show()
1
```

# 5. Box Plot

What it shows:

     Median
     Spread of data
     Outliers

Where to use:

     Detect unusual values
     Compare distributions

```python
sns.boxplot(data=data)
plt.title("Seaborn Box Plot")
plt.show()
```
